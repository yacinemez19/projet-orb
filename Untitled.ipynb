```python
from math import *
import numpy as np
import matplotlib.pyplot as plt

def creer_liste(doc):
    fichier = open(doc,'r')
    liste = fichier.read()
    fichier.close()
    liste = liste.split()
    for i in range(len(liste)):
        liste[i] = float(liste[i])
    return liste

def separer(liste,debut):
    piste = []
    for i in range(debut,len(liste),2):
        piste.append(liste[i])
    return piste
    
duo = []

t = creer_liste('instants.txt')
duo.append(creer_liste('duoI.txt'))
duo.append(creer_liste('duoII.txt'))
duo.append(creer_liste('duoIII.txt'))

solo = []
for i in range(3):
    solo.append(separer(duo[i],0))
    solo.append(separer(duo[i],1))

def trapeze(a,b,liste):
    assert type(a) == int
    assert type(b) == int
    
    return (b/2000-a/2000)*(liste[a]+liste[b])/2

def trapeze_composite(a,b,liste,N):
    assert type(a) == int
    assert type (b) == int
    
    S = 0
    h = int((b-a)/N)
    x0 = a
    x1 = a+h
    for i in range(N):
        S += trapeze(x0,x1,liste)
        x0 += h
        x1 += h
    return S

a = 0
b = len(t)
tau = 1

N = int(tau*(len(t)/10))

print(N)

t_crop = t[N+1 : 1-N]
```

```python
plt.plot(t_crop,solo[0][N+1 : 1-N])
trapeze(0,1000,solo[0])
```

```python
def S0moy(t):
    moy = (1/(2*tau))*trapeze_composite(t-N, t+N, solo[0], N)
    return moy

t_crop = t[N+1 : 1-N]

Y0 = [S0moy(i) for i in range(len(t_crop))]
```
